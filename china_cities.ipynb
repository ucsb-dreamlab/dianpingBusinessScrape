```python
import requests
import re 
import os 
import time
import traceback 
from bs4 import BeautifulSoup 
from lxml import etree
from fake_useragent import UserAgent
import pandas as pd
```

```python
def getHtmlContent(url):
 UA = UserAgent(verify_ssl = False)   
 headers = {
    "User-Agent": UA.random
    }

 try:
    s = requests.Session()
    r = s.get(url, headers = headers, cookies = {'from-my':'browser'})
    #r = s.get(url, headers = headers, cookies= s.cookies)
    r.raise_for_status()
    if(r.encoding != "UTF-8"):
        r.encoding = r.apparent_encoding
    return r.text
 except:
    return ""
```

```python
URL = 'http://www.dianping.com/citylist' #22 Jinan
html = getHtmlContent(url = URL)
soup = BeautifulSoup(html, "html.parser")
```

```python
cities_link = []
cities_name = []
for li in soup.find_all(class_= "letter-item"):
    for a in li.find_all(class_ = "link onecity"):
        cities_link.append(a.get('href').replace('//', '')+ "/ch75/g34311")
        cities_name.append(a.getText())
```

```python
import pandas as pd
Chinese_cities = pd.DataFrame({'city_name' : cities_name, 'URL' : cities_link})
Chinese_cities
```
