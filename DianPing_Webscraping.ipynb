```python
from fake_useragent import UserAgent
```

```python
def getHtmlContent(url, refer):
 UA = UserAgent(verify_ssl = False)   
 headers = {
    "User-Agent": UA.random,
    "Host": "www.dianping.com",
    "Referer": refer
    }

 try:
    r = requests.get(url, headers = headers)
    r.raise_for_status()
    if(r.encoding != "UTF-8"):
        r.encoding = r.apparent_encoding
    return r.text
 except:
    return ""
```

Retrieve the html code for a children's library in Jinan City:

```python
URL = 'https://www.dianping.com/shop/H9LR34sHvCC9Xn0h'
referer = 'https://www.dianping.com/search/keyword/22/0_%E7%BB%98%E6%9C%AC%E9%A6%86'
html = getHtmlContent(url = URL, refer=referer)
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup.prettify())
```
